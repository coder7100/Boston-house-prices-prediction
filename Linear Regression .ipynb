```python
import numpy as np
import pandas as pd
import matplotlib as mpl

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler 

from IPython.display import HTML
```

```python
#load the dataset
boston=load_boston()
#print the description of the dataset
#print(boston)
print(boston.DESCR)
```

```python
#put data into pandas dataframe

features=pd.DataFrame(boston.data,columns=boston.feature_names)
features
```

```python
targets=pd.DataFrame(boston.target,columns=["targets"])
targets
```

```python
df=pd.concat([features,targets],axis=1)
df
```

```python
df.describe().round(decimals=2)
```

```python
#lets create a corelation

corr=df.corr("pearson")

#take absolute value of corelation of the features and the targets

corrs=[abs(corr[attr]["targets"])for attr in list(features)]

#make a list of pairs(corr,features)

l=list(zip(corrs,list(features)))

#sort it in descending order using the values f\of corrs

l.sort(key=lambda x : x[0],reverse=True)
#we will no seprate it into 2 parameters.corrs and labels

corrs,labels=list(zip((*l)))

#plotting graph

index=np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index,corrs,width=0.5)
plt.xlabel("attributes")
plt.ylabel("correlation value")
plt.xticks(index,labels)
plt.show()
```

```python
#lets start normalisation
X=df["LSTAT"].values
Y=df["targets"].values
```

```python
#before normalisation 
Y[:5]
```

```python
x_scaler=MinMaxScaler()
X=x_scaler.fit_transform(X.reshape(-1,1))
X=X[:,-1]
y_scaler=MinMaxScaler()
Y=y_scaler.fit_transform(Y.reshape(-1,1))
Y=Y[:,-1]
```

```python
#after normalisation
Y[:5]
```

```python
#error function
def error(m,x,c,t):
    N=x.size
    e=sum(((m*x+c)-t)**2)
    return e*1/(2*N)
```

```python
#splitting the data

xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
```

```python
#update function

def update(m,x,c,t,learning_rate):
    grad_m=sum(2*((m*x+c)-t)*x)  
    grad_c=sum(2*((m*x+c)-t))
    m=m-grad_m*learning_rate
    c=c-grad_c*learning_rate
    return m,c
```

```python
def gradient_descent(init_m,init_c,x,t,learning_rate,iterations,error_threshold):
    m=init_m;
    c=init_c;
    error_values=[]
    mc_values=[]
    for i in range(iterations):
        e=error(m,x,c,t)
        if e<error_threshold:
            print("Error is less than threshold,gradient descent is stopped")
            break
        error_values.append(x)
        m,c=update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m,c,error_values,mc_values
    
```

```python
%%time
init_m=0.9
init_c=0
learning_rate=0.001
iterations=250
error_threshold=0.001
m,c,error_values,mc_values=gradient_descent(init_m,init_c,xtrain,ytrain,learning_rate,iterations,error_threshold)
```

```python
#making preps for animation
#we will take less no. of values to reduce the processing time
mc_values_anim=mc_values[0:250:5]
```

```python
#animation:


fig,ax=plt.subplots()
ln,=plt.plot([],[],"ro-",animated=True)

def init():
    plt.scatter(xtest,ytest,color="g")
    ax.set_xlim(0,1.0)
    ax.set_ylim(0,1.0)
    return ln,
def update_frame(frame):
    m,c=mc_values_anim[frame]
    x1,y1=-0.5,m*-.5+c
    x2,y2=1.5,m*1.5+c
    ln.set_data([x1,x2],[y1,y2])
    return ln,

anim=FuncAnimation(fig,update_frame,frames=range(len(mc_values_anim)),init_func=init,blit=True)
                   
HTML(anim.to_html5_video())
```

```python
#plotting the regression line

plt.scatter(xtrain,ytrain,color="r")
plt.plot(xtrain,((m*xtrain)+c),color="b")
```

```python
#plotting the error values

plt.plot(np.arange(len(error_values)),error_values,color="g")
plt.ylabel("error values")
plt.xlabel("iterations")
```

```python
#calculate the predicted values on xtest
predicted=(m*xtest)+c
```

```python
#calculation the mse
mean_squared_error(ytest,predicted)
```

```python
#make a dataframe for analysis
p=pd.DataFrame(list(zip(xtest,ytest,predicted)),columns=["xtest","ytest","predicted_values"])
p.head()
```
